## <a id='toc1_'></a>[Adding Titles and Axis Labels](#toc0_)

While we've touched on this topic before, it's worth revisiting and expanding on the essential skill of adding titles and axis labels to your plots. Clear, descriptive titles and labels are crucial for making your visualizations informative and easily understandable.


Let's start with a basic example and then explore some more advanced options:


````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
%matplotlib inline
````

````python
# Create some sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(6, 4))

# Plot the data
ax.plot(x, y)

# Add title and axis labels
ax.set(
    xlabel='X-axis',
    ylabel='Y-axis',
    title='Sine Wave'
)
````

In this example, we've used the `set_title()`, `set_xlabel()`, and `set_ylabel()` methods of the axis object to add our text elements. 


### <a id='toc1_1_'></a>[Customizing Titles and Labels](#toc0_)


You can customize these elements further:


````python
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(x, y)

ax.set_title('Sine Wave Over Time', fontsize=16, fontweight='bold')
ax.set_xlabel('Time (seconds)', fontsize=12, fontstyle='italic')
ax.set_ylabel('Amplitude', fontsize=12, color='blue')
````

Here, we've adjusted the font size, style, weight, and color of our text elements.


### <a id='toc1_2_'></a>[Multiple Lines in Titles and Labels](#toc0_)


You can also create multi-line titles or labels using `\n`:


````python
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(x, y)

ax.set_title('Sine Wave\nAmplitude vs. Time', fontsize=14)
ax.set_xlabel('Time\n(seconds)', fontsize=12)
ax.set_ylabel('Amplitude\n(meters)', fontsize=12)
````

Remember, clear and informative titles and labels are key to creating effective visualizations. They should concisely describe what the plot represents and provide context for the axes. In the next section, we'll delve deeper into customizing text properties to further enhance our plots.

## <a id='toc2_'></a>[Working with Text Properties](#toc0_)

Text properties in Matplotlib allow you to fine-tune the appearance of your text elements, making your plots more visually appealing and easier to read. Let's explore various ways to customize text properties.


### <a id='toc2_1_'></a>[Basic Text Properties](#toc0_)


Here's an example showcasing some common text properties:


````python
fig, ax = plt.subplots(figsize=(6, 4))

x = np.linspace(0, 10, 100)
y = np.sin(x)
ax.plot(x, y)

ax.set_title('Customized Text Properties', fontsize=16, fontweight='bold', color='navy')
ax.set_xlabel('X-axis', fontsize=12, fontstyle='italic', color='darkgreen')
ax.set_ylabel('Y-axis', fontsize=12, fontfamily='serif', color='darkred')
````

In this example, we've customized:
- `fontsize`: Size of the font
- `fontweight`: Weight of the font (e.g., 'normal', 'bold', 'light')
- `fontstyle`: Style of the font (e.g., 'normal', 'italic')
- `fontfamily`: Font family to use
- `color`: Color of the text


### <a id='toc2_2_'></a>[Using a Dictionary for Text Properties](#toc0_)


You can also use a dictionary to set multiple text properties at once:


````python
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(x, y)

title_props = {
    'fontsize': 16,
    'fontweight': 'bold',
    'color': 'navy',
    'verticalalignment': 'bottom'
}

ax.set_title('Title with Property Dictionary', **title_props)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
````

### <a id='toc2_3_'></a>[Text Alignment and Rotation](#toc0_)


You can control the alignment and rotation of text:


````python
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(x, y)

ax.set_title('Centered Title', ha='center')
ax.set_xlabel('Rotated X-Label', rotation=15)
ax.set_ylabel('Vertical Y-Label', rotation=0, ha='right')
````

Here, `ha` stands for horizontal alignment, and can be 'left', 'center', or 'right'.


### <a id='toc2_4_'></a>[Using LaTeX Rendering](#toc0_)


Matplotlib can render LaTeX expressions for high-quality typesetting:


````python
fig, ax = plt.subplots(figsize=(6, 4))

x = np.linspace(-10, 10, 100)
y = x ** 2

ax.plot(x, y)

ax.set(
    title='$y = x^2$',
    xlabel='X-axis',
    ylabel='Y-axis'
)
````

Note: LaTeX rendering requires a LaTeX installation on your system.


### <a id='toc2_5_'></a>[Custom Fonts](#toc0_)

## <a id='toc3_'></a>[Adding Annotations and Text Boxes](#toc0_)

Annotations and text boxes are powerful tools for highlighting specific data points or regions in your plots, providing additional context or explanations. Let's explore how to add these elements to your Matplotlib figures.


### <a id='toc3_1_'></a>[Basic Annotations](#toc0_)


Let's start with a simple annotation:


````python
fig, ax = plt.subplots(figsize=(6, 4))

x = np.linspace(0, 10, 100)
y = np.sin(x)
ax.plot(x, y)

ax.annotate(
    'Peak', xy=(np.pi/2, 1), xytext=(np.pi/2, 1.3),
    arrowprops=dict(facecolor='black', shrink=0.05),
    horizontalalignment='center'
)
````

Here, we've added an annotation pointing to the peak of the sine wave. The `xy` parameter specifies the point to annotate, and `xytext` determines where the text should be placed.


### <a id='toc3_2_'></a>[Customizing Annotations](#toc0_)


We can further customize our annotations:


````python
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(x, y)

ax.annotate(
    'Local minimum', xy=(3*np.pi/2, -1), xytext=(4, -0.5),
    arrowprops=dict(facecolor='red', shrink=0.05, width=2),
    fontsize=12, color='red',
    bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="red", lw=2)
)
````

In this example, we've customized the arrow color and width, text color and size, and added a box around the text.


### <a id='toc3_3_'></a>[Text Boxes](#toc0_)


Text boxes are useful for adding explanatory text to your plots:


````python
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(x, y)

ax.text(
    5, 0.5, 'This is a sine wave\nwith period 2π',
    bbox=dict(facecolor='white', edgecolor='black', alpha=0.7),
    ha='center', va='center'
)
````

The `text()` function allows you to add text at any location on your plot. The `bbox` parameter creates a box around the text.


### <a id='toc3_4_'></a>[Annotating Multiple Points](#toc0_)


You can annotate multiple points using a loop:


````python
fig, ax = plt.subplots(figsize=(6, 4))

x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)
ax.plot(x, y)

peaks = [(np.pi/2, 1), (5*np.pi/2, 1)]
for i, (xi, yi) in enumerate(peaks, 1):
    ax.annotate(
        f'Peak {i}', xy=(xi, yi), xytext=(xi, yi+0.3),
        arrowprops=dict(facecolor='black', shrink=0.05),
        horizontalalignment='center'
    )
````

Annotations and text boxes are excellent for drawing attention to specific aspects of your data or providing additional context. They can significantly enhance the interpretability of your plots when used judiciously. ## <a id='toc4_'></a>[Customizing Tick Labels](#toc0_)

Tick labels are an essential part of any plot, providing context and scale for your data. Matplotlib offers various ways to customize these labels to enhance readability and convey information more effectively. Let's explore some techniques for customizing tick labels.


### <a id='toc4_1_'></a>[Basic Tick Label Customization](#toc0_)


Let's start with a simple example of customizing tick labels:


````python
fig, ax = plt.subplots(figsize=(6, 4))

x = np.linspace(0, 10, 100)
y = np.sin(x)
ax.plot(x, y)

ax.set_xticks(np.arange(0, 11, 2))  # Set x-ticks at 0, 2, 4, ..., 10
ax.set_yticks(np.arange(-1, 1.1, 0.5))  # Set y-ticks from -1 to 1 in steps of 0.5

ax.tick_params(axis='both', which='major', labelsize=10, colors='navy')
````

Here, we've set custom tick locations and adjusted the size and color of the tick labels.


### <a id='toc4_2_'></a>[Rotating Tick Labels](#toc0_)


For long tick labels or to avoid overlapping, you can rotate them:


````python
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(x, y)

ax.set_xticks(np.arange(0, 11, 1))
ax.set_xticklabels(['Point {}'.format(i) for i in range(11)], rotation=45, ha='right')

plt.tight_layout()  # Adjust layout to prevent cut-off labels
````

### <a id='toc4_3_'></a>[Using Scientific Notation](#toc0_)


For very large or small numbers, scientific notation can be useful:


````python
fig, ax = plt.subplots(figsize=(6, 4))

x = np.linspace(1e5, 1e6, 100)
y = x**2
ax.plot(x, y)

ax.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
````

### <a id='toc4_4_'></a>[Custom Tick Formatter](#toc0_)


For more control, you can use a custom formatter:


````python
from matplotlib.ticker import FuncFormatter

def currency_formatter(x, p):
    return f'${x:,.0f}'

fig, ax = plt.subplots(figsize=(6, 4))

x = np.linspace(1000, 10000, 100)
y = x**2
ax.plot(x, y)

ax.yaxis.set_major_formatter(FuncFormatter(currency_formatter))
````

This example formats y-axis labels as currency.


### <a id='toc4_5_'></a>[Log Scale with Custom Labels](#toc0_)


For log scales, you might want to customize the labels:


````python
fig, ax = plt.subplots(figsize=(6, 4))

x = np.logspace(0, 3, 100)
y = x**2
ax.loglog(x, y)

ax.set_xticks([1, 10, 100, 1000])
ax.set_xticklabels(['1', '10', '100', '1K'])
ax.set_yticks([1, 1e2, 1e4, 1e6])
ax.set_yticklabels(['1', '100', '10K', '1M'])
````

### <a id='toc4_6_'></a>[Date Formatting](#toc0_)


For time series data, you can format ticks as dates:


````python
import matplotlib.dates as mdates
from datetime import datetime, timedelta

fig, ax = plt.subplots(figsize=(6, 4))

base = datetime(2023, 1, 1)
x = [base + timedelta(days=i) for i in range(180)]
y = np.random.randn(180).cumsum()
ax.plot(x, y)

ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.xticks(rotation=45)

plt.tight_layout()
````

This example shows monthly ticks for a 6-month period.


Customizing tick labels can greatly improve the readability and interpretability of your plots. By adjusting the format, rotation, and spacing of tick labels, you can ensure that your plot effectively communicates the scale and context of your data. ## <a id='toc5_'></a>[Practical Example and Best Practices](#toc0_)

In this final section, we'll explore some practical examples that combine various text, label, and annotation techniques we've learned. We'll also discuss best practices for using these elements effectively in your data visualizations.


Let's create a time series plot with annotations highlighting key events:


````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.dates as mdates


# Generate sample data
dates = pd.date_range(start='2022-01-01', end='2022-12-31', freq='D')
values = np.cumsum(np.random.randn(len(dates))) + 100

fig, ax = plt.subplots(figsize=(8, 4))

# Plot the data
ax.plot(dates, values)

# Customize the plot
ax.set_title('Stock Price Movement (2022)', fontsize=16, fontweight='bold')
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Price ($)', fontsize=12)

# Format x-axis ticks
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))

# Annotate key events
events = [('2022-03-15', 'Product Launch'), ('2022-07-01', 'Q2 Report'), ('2022-11-25', 'Black Friday')]
for date, event in events:
    idx = dates.get_loc(date)
    ax.annotate(event, (dates[idx], values[idx]),
                xytext=(10, 10), textcoords='offset points',
                arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.2'))

plt.tight_layout()
````

This example demonstrates how to create an informative time series plot with clear labels and annotations highlighting significant events.


### <a id='toc5_1_'></a>[Best Practices](#toc0_)


1. **Clarity is Key**: Ensure all text elements are easily readable. Use appropriate font sizes and colors that contrast well with the background.

2. **Consistency**: Maintain consistent styling (font, size, color) for similar elements across your plot.

3. **Avoid Clutter**: Don't overcrowd your plot with too many annotations or labels. Focus on the most important information.

4. **Use Hierarchy**: Make titles larger and bolder than axis labels, which should be more prominent than tick labels.

5. **Consider Your Audience**: Use terminology and units that your audience will understand.

6. **Proper Placement**: Position annotations and text boxes where they don't obscure important data points.

7. **Color Usage**: Use color to enhance readability, not just for aesthetics. Ensure sufficient contrast between text and background.

8. **Rotation and Alignment**: Rotate long axis labels if necessary, but try to keep them horizontal for easier reading when possible.

9. **Context is Crucial**: Provide enough context through titles, labels, and annotations for the plot to be understood standalone.

10. **Test for Accessibility**: Consider color-blind friendly palettes and check if your plot is understandable in grayscale.


By following these practices and combining the techniques we've learned, you can create informative, visually appealing, and professional-looking plots that effectively communicate your data insights. Remember, the goal is always to make your data easy to understand and interpret for your audience.
